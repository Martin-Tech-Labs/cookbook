# 05. Log semiring on graphs

This notebook visualizes paths with probabilities and their log-space interpretation.

```python

import math
import matplotlib.pyplot as plt

paths = {
    'A->B->D': [0.9, 0.5],
    'A->C->D': [0.6, 0.8],
    'A->E->D': [0.4, 0.95],
}

for name, probs in paths.items():
    joint = math.prod(probs)
    log_joint = sum(math.log(p) for p in probs)
    print(name, 'joint=', joint, 'log_joint=', log_joint)
```

```python

names = list(paths.keys())
linear = [math.prod(p) for p in paths.values()]
logvals = [sum(math.log(p) for p in probs) for probs in paths.values()]

fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].bar(names, linear)
axes[0].set_title('Path probabilities')
axes[0].tick_params(axis='x', rotation=30)
axes[1].bar(names, logvals)
axes[1].set_title('Path log probabilities')
axes[1].tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.show()
```

## Log semiring idea

In probability space, paths multiply along edges and add across alternatives. In log space, multiplication becomes addition, which is often easier and more stable.
