# 04. Log-space joint probability

This notebook explains why log space is so useful for sequence probabilities.

```python

import math

probs = [1e-5, 1e-6, 1e-4, 1e-3]
product = 1.0
for p in probs:
    product *= p
log_sum = sum(math.log(p) for p in probs)
recovered = math.exp(log_sum)

print('direct product:', product)
print('sum of logs:', log_sum)
print('exp(sum of logs):', recovered)
```

```python

seq_a = [-0.2, -0.4, -0.1]
seq_b = [-0.8, -0.9, -0.3]

print('sequence A total logprob:', sum(seq_a))
print('sequence B total logprob:', sum(seq_b))
print('A more likely than B?', sum(seq_a) > sum(seq_b))
```

## Why this matters

Comparing sequences by summed logprob is numerically stable and preserves the same ordering as multiplying probabilities.
