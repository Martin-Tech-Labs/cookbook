# 07. A custom algebra system

We define a custom binary operation

$$a \oplus b = a + b + rac{ab}{4}$$

and study how this interaction rule differs from ordinary addition.

```python

import numpy as np
import matplotlib.pyplot as plt

def op(a, b):
    return a + b + (a * b / 4)
```

## Basic properties

```python

values = [(-2, 3), (1, 4), (0.5, 0.25)]
for a, b in values:
    print((a, b), 'commutative?', np.isclose(op(a, b), op(b, a)))
```

```python

def assoc_error(a, b, c):
    return op(op(a, b), c) - op(a, op(b, c))

triples = [(1,2,3), (0.5, 1.5, 2.0), (-1, 2, 4)]
for triple in triples:
    print(triple, 'associativity error:', assoc_error(*triple))
```

```python

# identity e satisfies a ⊕ e = a
# a + e + ae/4 = a  => e(1 + a/4) = 0
# for generic a, e = 0
print('identity element candidate: 0')
```

## Compare with standard addition

```python

a_vals = np.linspace(-4, 4, 200)
b_fixed = 2
standard = a_vals + b_fixed
custom = op(a_vals, b_fixed)

plt.figure(figsize=(6,4))
plt.plot(a_vals, standard, label='a + b')
plt.plot(a_vals, custom, label='a ⊕ b')
plt.legend()
plt.title('Custom algebra vs ordinary addition')
plt.xlabel('a')
plt.ylabel('result')
plt.grid(True)
plt.show()
```

## Repeated composition

```python

def repeated_op(a, n):
    value = a
    out = [value]
    for _ in range(n - 1):
        value = op(value, a)
        out.append(value)
    return out

seq = repeated_op(1.0, 8)
print(seq)

plt.figure(figsize=(6,4))
plt.plot(range(1, len(seq)+1), seq, marker='o')
plt.title('Repeated application of a ⊕ a ⊕ a ...')
plt.xlabel('number of terms')
plt.ylabel('value')
plt.grid(True)
plt.show()
```

## Conceptual connection

This is **not** a probability system. It is a custom algebra system with its own interaction rule. The point is that changing the algebra changes how quantities combine. That is similar in spirit to working in log space, where multiplication becomes addition, but the interpretation here is purely algebraic rather than probabilistic.
