# 06. Decibels and logarithmic scales

This notebook explains why logarithmic scales are useful, what decibels measure, and how to build intuition for decibel arithmetic.

A central idea is that **decibels turn multiplication of ratios into addition of scores**. That makes them easier to reason about when the raw linear scale spans huge ranges.


## Why decibels were introduced

In many physical and engineering systems, linear ratios can vary across enormous ranges. For example, a power ratio might be 0.001, 1, 1000, or much larger. Writing and comparing those values directly on a linear scale is awkward.

A logarithmic scale compresses those large ranges into a more manageable form. Decibels are one standard way to do that.


```python
import numpy as np
import matplotlib.pyplot as plt

power_ratios = np.logspace(-3, 3, 200)
amplitude_ratios = np.logspace(-3, 3, 200)

db_power = 10 * np.log10(power_ratios)
db_amplitude = 20 * np.log10(amplitude_ratios)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,4))
axes[0].plot(power_ratios)
axes[0].set_title('Linear power ratios')
axes[1].semilogy(power_ratios)
axes[1].set_title('Same values on log y scale')
axes[2].plot(db_power)
axes[2].set_title('Power ratios in dB')
plt.tight_layout()
plt.show()
```

## Core formulas

For **power ratios**:

$$\text{dB} = 10 \log_{10}(\text{power ratio})$$

For **amplitude ratios** (like voltage or pressure, when power is proportional to amplitude squared):

$$\text{dB} = 20 \log_{10}(\text{amplitude ratio})$$


```python
examples = [0.5, 2, 10, 100]
for ratio in examples:
    print('power ratio', ratio, '->', 10 * np.log10(ratio), 'dB')
    print('amplitude ratio', ratio, '->', 20 * np.log10(ratio), 'dB')
    print('---')
```

## Important intuition: multiplying ratios becomes adding decibels

This is why decibels are convenient. If one stage doubles power and another stage also doubles power, then the total power ratio is:

$$2 \times 2 = 4$$

In decibels, that is:

$$10 \log_{10}(4) \approx 6.02\,\text{dB}$$

Since one doubling of power is about **+3.01 dB**, two doublings add up:

$$3.01 + 3.01 \approx 6.02\,\text{dB}$$

So in decibel language, multiplication in the original ratio space becomes addition in dB space.


```python
db_of_2 = 10 * np.log10(2)
db_of_4 = 10 * np.log10(4)
db_of_10 = 10 * np.log10(10)

print('2x power  ->', db_of_2, 'dB')
print('4x power  ->', db_of_4, 'dB')
print('10x power ->', db_of_10, 'dB')
print('2x then 2x in dB ->', db_of_2 + db_of_2, 'dB')
```

### Rules of thumb

For **power ratios**:
- doubling is about **+3 dB**
- multiplying by 10 is **+10 dB**

Those are rough but very useful mental anchors.


## About expressions like `40 dB` combined with `-20 dB`

This should be handled carefully. Decibel values are already logarithmic summaries of underlying ratios. So multiplying dB numbers directly is **not** the standard physical rule.

The meaningful standard operations are:
- **add dB values** when multiplying underlying ratios
- **subtract dB values** when dividing underlying ratios

For example, a +40 dB ratio combined with a -20 dB ratio results in a +20 dB ratio because in dB space that becomes addition:

$$40 + (-20) = 20$$

So the meaningful interpretation is that the underlying linear ratios are multiplied while the dB values are added.


```python
db_a = 40
db_b = -20
combined_db = db_a + db_b

power_ratio_a = 10 ** (db_a / 10)
power_ratio_b = 10 ** (db_b / 10)
combined_ratio = power_ratio_a * power_ratio_b
combined_ratio_db = 10 * np.log10(combined_ratio)

print('40 dB corresponds to power ratio:', power_ratio_a)
print('-20 dB corresponds to power ratio:', power_ratio_b)
print('combined ratio:', combined_ratio)
print('combined in dB by addition:', combined_db)
print('combined in dB from ratio:', combined_ratio_db)
```
