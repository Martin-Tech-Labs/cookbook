# 02. Cross-entropy and perplexity

This notebook connects prediction quality to negative log likelihood, cross-entropy, average logprob, and perplexity.

## Manual intuition

If the model assigns high probability to the correct token, the negative log likelihood is small. If it assigns low probability, the loss is large.

```python

import math

good_prediction_prob = 0.8
bad_prediction_prob = 0.05

print('good NLL:', -math.log(good_prediction_prob))
print('bad NLL:', -math.log(bad_prediction_prob))
```

```python

example_logprobs = [-0.1, -0.3, -0.2, -0.4]
avg_logprob = sum(example_logprobs) / len(example_logprobs)
perplexity = math.exp(-avg_logprob)

print('average logprob per token:', avg_logprob)
print('perplexity:', perplexity)
```

## Cross-entropy from logits

Cross-entropy is a standard training objective for classification and language modeling.

```python

import torch

logits = torch.tensor([[2.5, 0.3, -1.2], [0.1, 0.2, 2.1]])
targets = torch.tensor([0, 2])
loss = torch.nn.functional.cross_entropy(logits, targets)
print('cross-entropy loss:', loss.item())
```

## Good vs bad prediction comparison

```python

probs_good = torch.tensor([0.9, 0.08, 0.02])
probs_bad = torch.tensor([0.2, 0.3, 0.5])
correct_index = 0

nll_good = -torch.log(probs_good[correct_index]).item()
nll_bad = -torch.log(probs_bad[correct_index]).item()

print('good prediction NLL:', nll_good)
print('bad prediction NLL:', nll_bad)
print('good perplexity contribution:', math.exp(nll_good))
print('bad perplexity contribution:', math.exp(nll_bad))
```
